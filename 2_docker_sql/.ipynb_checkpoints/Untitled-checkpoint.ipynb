```python
df=pd.read_parquet('yellow_tripdata_2021-01.parquet', engine='pyarrow')
```

```python
from sqlalchemy import create_engine
engine = create_engine(r"postgresql://root:root@localhost:5432/ny_taxt")
```

```python
engine.connect()
```

```python
print(pd.io.sql.get_schema(df, name = "yellow_taxi_data", con = engine))
```

```python
df.to_csv("yellow_taxi_data.csv")
```

```python
df_iter = pd.read_csv("yellow_taxi_data.csv", iterator=True, chunksize = 1000)
```

```python
df.to_sql(name="yellow_taxi_data", con = engine, if_exists = "replace")
```

```python
df = next(df_iter)
```

```python
df.drop(columns=["Unnamed: 0"], inplace = True)
```

```python
df.head(n=0).to_sql(name="yellow_taxi_data", con = engine, if_exists = "replace")
```

```python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime )
```

```python
df.to_sql(name="yellow_taxi_data", con = engine, if_exists = "replace")
```

```python
from time import time
while True: 
    time_start = time()
    
    df = next(df_iter)
    df.drop(columns=["Unnamed: 0"], inplace = True)
    df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
    df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime )
    
    df.to_sql(name="yellow_taxi_data", con = engine, if_exists = "append")
    time_end = time()
    
    print("inserted another chunk, toook %.3f second" % (time_end - time_start))
```
